```python
import random
from operator import mul
from functools import reduce
import pdb



# pretty print a table
def print_table(table):

    # print in match id order
    for i in sorted(table.keys()):
        print(f"id: {table[i]['id']}\tmatch_start:{table[i]['match_start']}\tval:{table[i]['val']}")




# mock choices
first_start = 1600
match_starts = [1600, 2200]
vals = [1,2,3]

# generate mock table
table = {}
for i in range(13):

    # create fake row
    row = {'id':i, 'val':random.choice(vals), 'match_start':random.choice(match_starts)}
    table[i] = row


# get list of vals to calculate system
table_vals = [ x['val'] for x in table.values() ]

# print system
print(f"Org system:\t{reduce(mul, table_vals)}")
print_table(table)
print("\n\n")


### adjust the table so that matches that start first are prioritized to get 1's

# find all first starts that are not 1's
first_matches_not_spik = {}
late_matches_spik = {}
for key,match in table.items():

    # if it's a first match
    if match['match_start'] == first_start:

        # if it is not a 1
        if match['val'] != 1:

            # save it
            first_matches_not_spik[match['id']] = match


    # find all later starts that are 1's
    else:
        # if it is a 1
        if match['val'] == 1:

            # save it
            late_matches_spik[match['id']] = match



# make all possible first starts 1's

# go through the matches in a random order
match_list = list(first_matches_not_spik.items())
random.shuffle(match_list)
for key,match in match_list:

    # check if there are no more late matches to switch with
    if len(late_matches_spik) == 0:
        # exit the loop
        continue


    # pick a random late match to switch with
    random_late_match_key, random_late_match = random.choice(list(late_matches_spik.items()))

    # switch vals
    table[random_late_match_key]['val'] = match['val']
    table[key]['val'] = 1

    # remove both matches from their respective lists
    del first_matches_not_spik[key]
    del late_matches_spik[random_late_match_key]


# get list of vals to calculate system
table_vals = [ x['val'] for x in table.values() ]

# print system
print(f"Mod system:\t{reduce(mul, table_vals)}")
print_table(table)
```
